```python
import matplotlib.pyplot as plt
import matplotlib.colors as plt_colors
import numpy as np
```

```python
f = open('img.ppm','rb')
lc = 0
header = []
while(lc != 3):
    line = f.readline().decode()
    header.append(line)
    if line[0] != '#':
        lc += 1
print("header of the file: \n",header)
h,w = [int(a) for a in header[-2].split(sep = " ")]
#print(h,w)
cdepth = int(header[-1])
#print(cdepth)
#find the number of bytes in the file
num_bytes = 2
if cdepth == 255:
    numbytes = 1
    
buffer = f.read(h*w*num_bytes)

dt = np.uint16
if num_bytes == 1:
    dt = np.uint8

array = np.frombuffer(buffer,dtype=dt)
array = array.reshape((h,w)).T
f.close()
```

```python
fig = plt.figure(figsize = (20,20), dpi = 200)
ax = plt.Axes(fig,[0,0,1,1])
ax.set_axis_off()
ax.imshow(array, norm=plt_colors.LogNorm(), cmap = plt.get_cmap('cubehelix'))
fig.add_axes(ax)
plt.savefig("figure.png")
```
